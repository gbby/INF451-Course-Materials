```python
# https://open.toronto.ca/dataset/covid-19-cases-in-toronto/

import pandas as pd
import requests
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns

# set the standard plot size
plt.rcParams['figure.figsize']=(15,15)
sns.set()
```

```python
# Get the dataset metadata by passing package_id to the package_search endpoint
# For example, to retrieve the metadata for this dataset:

url = "https://ckan0.cf.opendata.inter.prod-toronto.ca/api/3/action/package_show"
params = { "id": "64b54586-6180-4485-83eb-81e8fae3b8fe"}
package = requests.get(url, params = params).json()
# print(package["result"])
```

```python
# Get the data by passing the resource_id to the datastore_search endpoint
# See https://docs.ckan.org/en/latest/maintaining/datastore.html for detailed parameters options
# For example, to retrieve the data content for the first resource in the datastore:

for idx, resource in enumerate(package["result"]["resources"]):
    if resource["datastore_active"]:
        url = "https://ckan0.cf.opendata.inter.prod-toronto.ca/api/3/action/datastore_search"
        p = { "id": resource["id"] }
        data = requests.get(url, params = p).json()
        df = pd.DataFrame(data["result"]["records"])
        break
        
# CKAN stands for Comprehensive Knowledge Archive Network
```

```python
# let's try seaborn, a declarative front-end to matplotlib
# this distribution plot is basically a histogram-like method, but we'll cheat a bit by using it since seaborn's default bar method wants averages

g = sns.displot(df, x="Source of Infection")
g.set_xticklabels(rotation=90)
g.fig.set_size_inches(15,15) # and we'll use this to change the figure size
```


# Something to keep in mind...

There is a difference between bar charts and histograms. With bar charts, a column represents a group defined by a categorical variable. With histograms, a column represents a group defined by a continuous, quantitative variable. I'm being a bit pedantic here, as the methods I'm using here really only come down to whether or not I've re-shaped my original dataframe. (I could "follow the rules" by simply creating a new dataframe with a column for age categories and a column for their respective counts, but we can cheat by using histogram methods.) Having a sense of what kind of visual output I'm aiming for may determine how much time I spend re-shaping my original tabular/spreadhsheet data.


```python
# anyway, let's do some comparisons. 
# matplotlib doesn't handle multivariate visualization very easily, so we'll use seaborn's "dodge" option

g = sns.displot(df, x="Source of Infection", hue="Client Gender", multiple="dodge")
g.set_xticklabels(rotation=90)
g.fig.set_size_inches(20,10)
```

```python
# side-by-side comparison is also easy
g = sns.displot(df, x="Source of Infection", hue="Client Gender", multiple="dodge", col="Client Gender")
g.set_xticklabels(rotation=90)
sns.set(font_scale=1.3) # seaborn's method for setting fontsize is pretty dumb
g.fig.set_size_inches(20,10)
```

```python
# seaborn's countplot method is a good alternative to displot
# we can use it to show the counts of observations in each categorical bin
# remember, that seaborn's barplot method, by default, sets bar height to averages
plt.figure(figsize=(20,10))
g = sns.countplot(data=df, x='Source of Infection', palette='colorblind')
```


# What's Missing?

Are any of you wondering why the counts are so low? By default, CKAN's default download is 100 rows (and its upper limit is 32000). There have been over 17000 cases of COVID-19 in Toronto, so the visualizations we've created up to now would be useless! Be careful with your data sources, especially if the API is provided with little documentation!! Of course, we could have avoided this mess by downloading the .csv and using pd.read_csv() to create our dataframe, but we'd run the risk that our data would get stale.

Anyway, let's go back up to cell 3 and add ?limit=32000 to the end of the url so that it looks like this: https://ckan0.cf.opendata.inter.prod-toronto.ca/api/3/action/datastore_search?limit=32000


```python
# now re-import the data 
for idx, resource in enumerate(package["result"]["resources"]):
    if resource["datastore_active"]:
        url = "https://ckan0.cf.opendata.inter.prod-toronto.ca/api/3/action/datastore_search?limit=32000" # with updated url
        p = { "id": resource["id"] }
        data = requests.get(url, params = p).json()
        full_df = pd.DataFrame(data["result"]["records"]) # and make sure you give your full dataframe a different variable name so you don't run into state issues
        break
```


# Pay attention to the legend on the right as we update to the full dataset. Notice anything different?


```python
g = sns.displot(full_df, x="Source of Infection", hue="Client Gender", multiple="dodge") # notice that I updated the data source
g.set_xticklabels(rotation=90)
g.fig.set_size_inches(20,10)
```


# And what else might we consider visualizing while we're at it? Maybe age group by gender...

```python
# or how about age group x outcome
g = sns.displot(full_df, x="Age Group", hue="Outcome", multiple="dodge")
g.set_xticklabels(rotation=90)
g.fig.set_size_inches(20,10)
```

```python
# and let's add some interaction using altair
# like seaborn, it's a simple declarative front-end that makes it really easy to switch options/parameters
# also has a great built-in chart editor tool

import altair as alt

alt.Chart(df).mark_bar().encode( # notice that I switched back to df from full_df... altair has a 5000 row max
    alt.X("Age Group"),
    y='count()',
    color='Source of Infection', # could switch between gender, outcome, neighbourhood, source of infection, etc...
    tooltip=['Age Group', 'Client Gender', 'Outcome']
).interactive().properties(
    width=600, height=400
).configure_axis(
    labelFontSize=18,
    titleFontSize=18
)
```
