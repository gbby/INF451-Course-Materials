<p>William Playfair (1759 â€“ 1823) was a Scottish engineer, political economist, secret agent on behalf of Great Britain during its war with France, and a data visualization pioneer. He is said to have invented several graphic types - in 1786, the line, area, and bar chart of economic data, and in 1801, the pie chart and circle graph. The image below is one of his most famous graphics, and is recognized as possibly the first ever bar chart depicting statistical data.</p> 

<p><img style="float: left;margin:5px 20px 5px 1px" src="playfair_wheat.png"></p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Use pandas to read in the dataset
# Data comes from the R HistData package: https://rdrr.io/cran/HistData/man/HistData-package.html
# Thank Michael Friendly (who we read in Week 1) for providing it
df = pd.read_csv("data/wheat.csv")
```

```python
df.tail()
```

```python
fig, ax = plt.subplots()

# background image
img = plt.imread("playfair_cropped.png")
ax.imshow(img, extent=[-2.5, 55, -3, 105], aspect = 'auto')
ax.set(ylim=(0, 100))

# bars
ax1 = sns.barplot(x="Year", y="Wheat", data=df, color='grey', alpha=0.6)
ax1.set(ylim=(0, 100))

# line
ax2 = ax1.twinx()
ax2.plot(ax1.get_xticks(), df.Wages, color='red')
# ax2.set(ylim=(0, 100)) # check out that 0-100 range that playfair uses! comment out this line to see what matplotlib's default would be...

fig.set_size_inches(24, 12)
```
