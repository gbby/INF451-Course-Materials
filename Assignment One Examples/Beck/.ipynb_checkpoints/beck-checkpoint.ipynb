# A Redesign of Harry Beck's Tube Map

## Background: Harry Beck's Tube Map
<p>Harry Beck was an Engineering draughtsman who was famous for designing a radically different version of the tube map in 1933 that became iconic (Transport for London, 2019). The 
underlying principles that Beck used for his rendition of the London Underground have persisted today, and we can see that when comparing the 1933 map (left) to today's Tube map 
(right) which is accurate as of May 2019.<p>

<img style="float: left; margin: 10px 10px 10px 0; max-width:600px" src="beck_map.jpg">
<img style="margin: 10px 0 0 10px; max-width:600px" src="https://tfl.gov.uk/cdn/static/cms/images/tube-map.gif">

<p>Beck's map demonstrates a more abstract, topological intepretation of London that presents the tube station connections in a way that is not accurate in terms of 
distance and location, however it is memorable, attractive and fairly easy to make sense of at a high level. The diagram's core is stretched out and much easier to 
digest than if we were comparing it directly to the true-to-size core of the city as seen in a map on paper or on Google Maps. The modern Tube Map maintains Beck's 45 degree angles, 
parallel lines, use of colour and abstract, skewed representation of the subway lines.</p>


## Critique of Beck's Design
<p>Beck's design - as well as the official designs inspired by it since - has become so ingrained into Londoners' experience of the city that it heavily influences how they perceive 
and experience London (Vertesi, 2008). Deviations from Beck's norm were met with distain, and the above ground has instead become the distorted version of the city; wayfinding is in many cases 
understood by locating where one is in respect to the Tube map (Vertesi, 2008).</p>

<p>It is daunting to suggest a redesign of an artifact that is so revered and heavily publicised, however while Beck's map has done a good job of reshaping Londoners' 
views of London, this critique is more about creating a version of the map that Londoners (new visitors and those who live and work in London) might be able to use more 
easily to actually get where they need to go. One interviewee in Vertesi's paper about the London tube map noted that, due to its distortion, they had no idea that they 
lived so close to their workplace and could have walked instead. Others with less experience with the city <i>relied</i> on the Tube map to get around rather than take 
the above-ground options that more experienced people were able to (Vertesi, 2008). </p>
    
<p>My critique takes on a humanistic approach, focusing on the visitor to London and their goal - to get from Point A to Point B in london using the Tube. Since my redesign 
uses accurate distances between stops, a benefit of this is that if the visitor doesn't need to go far, the map will show them that they are in walking or biking distance. 
They may not need to take the Tube at all, and so this may help to educate newer Londoners about how the city is truly laid out so they will have more options for travel in 
the future. The aim of this redesign is to make the wayfinding within the city more personalized and accessible to more people (especially newcomers). My perspective is limited 
as I have only ridden the tube once, and so my redesign should be taken as a first iteration. If this design is to be explored more in the future, perspectives from more people 
who have visited or live in London would need to be taken into account to make the map better.</p>

### This redesign focuses on two critical aspects which I address with my redesign:
<p><ol>

<li>Usability and Accessibility: Beck's map, as well as the modern map, is large, complicated and potentially overwhelming to look at for those unfamiliar with the map and city.
The map could be more simple while still preserving the layout of the city and showing Londoners where they need to go. In addition, the legend for the various map lines is tucked away 
at the bottom, forcing people to drag their attention back and forth between the map and legend if they are unsure what the different lines and colours represent.</li>

<li>Goal-Oriented Design: It can be argued that Beck's map somewhat prioritizes form over function - it was iconic and still is, but it distorts what London looks 
like which does not teach visitors and Londoners to find their way above ground. It also omits certain areas of the line to conserve space (e.g. the District line), which may obscure the viewer's destination. 
The map could benefit by being more focused on the wayfinding goals of newcomers and Londoners alike who don't know what London looks like above ground. Viewing the map this way 
would hopefully also educate Londoners about the size and scale of the city so they will be better equipped to navigate it <i>outside</i> the Tube as well.</li>
</ol></p>

<p>Beck's map has been called a "network diagram" (Vertesi, 2008) rather than a map due to its removal from representing geography. My critique does not assume that his representation
is not usable, it simply explores another interpretation of the map that uses modern technology. The intent of this redesign is that each Tube line in London would have a screen or kiosk that contains this 
new, interactive map. It could not have existed in 1933, however this redesign is meant to be merely an exploration.</p>

## Data Used

<p>The data used for this redesign was retrieved from <a href="https://github.com/oobrien/vis/tree/master/tube/data">https://github.com/oobrien/vis/tree/master/tube/data</a>. 
Using this, I was able to plot out the London Tube lines and stations, and use it as a basis for the redesign (see "London Tube Stations" and "London Tube Lines" below). On its own without any 
refinement, it is clear that the tube lines are clustered and hard to interpret, especially towards the centre. We can also see how drastically different Beck's interpretation of the map was, 
and how much work he put into reorganizing it.</p>

```python
#Read files
import geopandas as gpd
stations = gpd.read_file('data/stations.shp')
lines = gpd.read_file('data/lines.shp')

#Plot the data
%matplotlib inline
station_plot = stations.plot(figsize=(14, 14))
lines_plot = lines.plot(figsize=(14, 14)) 

station_plot.plot()
lines_plot.plot()

#Add titles to plots
station_plot.set_title("London Tube Stations", fontsize=20)
lines_plot.set_title("London Tube Lines", fontsize=20)

#Map data retrieved from https://github.com/oobrien, trimmed using QGIS to show just underground lines and stations and then converted to shape files.
```

## The Redesign Process

### Step 1: Start with an Accurate Map
<p>The Tube data that I retrieved contained all of the modern stations and above-ground trains, and so it was my starting point. I imported the data 
into QGIS, removed all of the above-ground lines and stations, then brought into Illustrator so I could alter it easily. Using a 
"true" map of London with accurate distances will provide Londoners with an accurate view into the city, and make it easier to understand its complexity.</p>
<img style="margin: 10px; max-width:800px" src="tube-map-prototype1.png">

### Step 2: Distinguish the Lines with Colour
<p>Next, using <a href="https://www.arcgis.com/home/webmap/viewer.html?webmap=8a567ebac15748d39a747649a2e86cf4">this map</a> as a guide, I separated the different lines using colour.
This distinction is necessary to make the map more usable and understandable, and allow Londoners to know which line is which.</p>
<img style="margin: 10px; max-width:800px" src="tube-map-prototype2.png">

### Step 3: Remove Modern Stations and Lines
<p>I wanted the map to be compared closely with Beck's map, so I removed by hand all lines and stations that were not available in 1933 and did not appear on Beck's original map. This 
representation may not be perfect due to its complexity, however it should be fairly close, and good enough for exploratory purposes.</p>
<img style="margin: 10px; max-width:800px" src="tube-map-prototype3.png">

### Step 4: Add Names to the Lines
<p>In order to enhance usability and accessibility, I added names for each line with large, clear and colour-coded text to the ends of each line. Ideally, this would 
help Londoners quickly find the specific line they need to look at without having to search back and forth between the legend and map. The large text will also help with those 
with vision difficulties, or older individuals.</p>
<img style="margin: 10px; max-width:800px" src="tube-map-prototype4.png">

### Step 5: Simplify the Legend
<p>Since this geographically accurate version of the Tube map is more complex, I found the names of the lines too large and eye-catching, as they take away from the rest of the map.
To address this, I shrunk them to only display the first two letters of the line name. For those that don't know the names of the lines, I also added a simple colour-coded legend 
with large text to the bottom of the map. This way, Londoners will not have to only rely on colour or just the legend - they will be able to recognize the line name by these two-letter guides 
that are placed close to the lines themselves.</p>

### Step 6: Add a Goal-Focused Wayfinding Method
<p>By using an accurate layout of London, this redesign is too large and complex to show all the line names immediately. To assist Londoners find their way, a simple interactive element is 
introduced. Londoners' current location is indicated on the Current Location on the top right as well as on the map itself, shown by an emphasized station dot.
This map, which is intended to be presented at each Tube station on at least one screen, prompts its viewer to enter their destination in the large text box. They will (ideally) be able to type 
out their destination either by station name, address or attraction name, and the map will display the closest station to that destination.</p>
<img style="margin: 10px; max-width:800px" src="tube-map-flow-1.png">

### Step 7: Show Only What They Need to See 
<p>After the Londoner enters their destination, the map will zoom in and narrow down on <i>only the relevant tube lines</i> based on where they are trying to go. In this example, 
The map viewer wants to get from their current location of Stonebridge Park to British Museum (this station has since closed, but existed on Beck's map). The map 
also shows the station that the viewer must transfer at. At this stage, only three stations are shown which is likely the bare minimum that the viewer will need to see to get where 
they need to go.</p>
<img style="margin: 10px; max-width:800px" src="tube-map-flow-2.png">

### Step 8: Provide a Deeper Look Into the Map
<p>Finally, if the Londoner is confused or wishes to see the stations around their destination on the map, they can zoom in via the well-known pinch-and-zoom interaction to get a 
closer look into the surrounding area. Not all stations will show by default, however if the viewer taps a station, it will become highlighted and show the name (e.g. "Notting Hill Gate"). 
The viewer's current location will still be pointed to off the map, so they do not lose a sense of where they are.</p>
<img style="margin: 10px; max-width:800px" src="tube-map-flow-3.png">

## Conclusions and Next Steps

<p>The redesign outlined above is based on real data and station locations, however it is still speculative. Beck's map was printed, and this one requires many touch-screens, which would be 
impossible in 1933 and very expensive today. It might not be feasible, but it explores what an alternate view into his map may have looked like if this technology was available. It certainly 
would have changed the way that Londoners consumed maps because it was interactive rather than static, and it may have changed the way that maps in other areas of the world have been 
designed since. If we always had the ability to focus on only the stations and locations relevant to our trip, would the principles pushed by Beck's classic, stylized yet distorted 
version of the map still perservere today?</p>

<p>I believe the goals (help people navigate the tube) and audiences (visitors of London and people who live there) of Beck's map and my redesigned map are the same, however my approach 
focuses on a more human-centric lens, and uses modern technology. In summary, here are the ways that my redesigned map attempts to address the critical points outlined earlier:
    
<p><b>Usability and Accessibility</b></p>
<ul>
<li>The map uses large, colour-coded text in the legend and displays line names close to the lines so looking back and forth is not necessary.</li>
<li>Station names are cut from view to make the map less overwhelming; depending on where the viewer is going, they only see where they are, where they must transfer and where they need to be.</li>
<li>When the viewer enters their destination, they only see the Tube lines they need to care about. This "chunking" of the map makes it more approachable for newcomers.</li>
</ul>

<p><b>Goal-Oriented Design</b></p>
<ul>
<li>The map is interactive and personalized, and provides Londoners with a focused view into where they need to go by prompting them for their destination.</li>
<li>It provides a realistic view of the city and underground so new Londoners can more easily familiarize themselves with the size and layout of London.</li>
<li>Ideally, the map would allow Londoners to enter a destination that is an address, point of interest or station, and display the closest station. This would greatly reduce the time it takes for someone to find out how to get there.</li>
</ul>

There are a few additions that this redesigned map would benefit from but were not added due to time constraints. One is the location of the River Thames. This redesign intends to make 
overall wayfinding in London easier, however it leaves out an important landmark. In the next iteration, the map should include the river. Additionally, this redesign intends to improve accessibility, 
yet only focuses on precise touch interactions to navigate. More consideration is needed around alternate ways to input information, as well as the size and location of the screens. Lastly, the colours 
of the lines themselves could be refined further to account for colour blind users by adjusting brightness, or introducing patterns to the lines to differentiate them.
